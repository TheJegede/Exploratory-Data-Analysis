### Problem 1 - Car Dataset

```python
import pandas as pd
```

```python
mtcars = pd.read_csv('carsDataset.csv')
```

```python
mtcars
```

```python
mtcars.set_index('cars.features', inplace=True)
```

```python
mtcars
```

```python
# Problem 1: Find outliers for mpg, qsec, and hp for cars data
print("\n2. Calculating outliers for target columns, using anything above/below 2 standard deviations.")

# Parameters
factor = 2
Cols = ['mpg', 'qsec', 'hp']

outliers = {} 

# Loop through each column
for col in Cols:
    mean = mtcars[col].mean()  # Calculate mean
    std = mtcars[col].std()    # Calculate standard deviation
    
    upper_lim = mean + std * factor  # Upper limit
    lower_lim = mean - std * factor  # Lower limit
    
    # Find outliers
    outliers[col] = mtcars[(mtcars[col] >= upper_lim) | (mtcars[col] <= lower_lim)]
    
    # Print outliers for the column
    print(f"\nThe outliers for {col} are:\n{outliers[col]}\n")
```

### Problem 2 - Bank Loans

```python
BankLoans = pd.read_csv('bankloan.csv')
```

```python
BankLoans
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import DBSCAN
```

```python
#Print preview of dataframe
print('Bank Loan Data')
BankLoans.head()
```

```python
# Columns to plot
Cols = ['x1', 'x5', 'x6', 'x7', 'x11', 'x13', 'x14']

# Create a grid of subplots (3 rows, 3 columns)
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 10))
fig.suptitle("Boxplots for Bank Loan Data", fontsize=16)

# Flatten axes array to iterate easily (handles cases with fewer plots than subplots)
axes = axes.flatten()

# Generate boxplots
for i, col in enumerate(Cols):
    BankLoans[col].dropna().to_frame().boxplot(ax=axes[i])
    axes[i].set_title(f"Boxplot for {col}")

# Remove empty subplots if the number of plots is less than 9
for j in range(len(Cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to fit titles
plt.show()
```

```python
#scale the data to calculate distances
scaler = MinMaxScaler()
bankLoanSeries = scaler.fit_transform(BankLoans.fillna(0))
```

```python
#Create clusters from dataset 
outlier_detection = DBSCAN(min_samples=2, eps=1.13)
clusters = outlier_detection.fit_predict(bankLoanSeries)
```

```python
print("\n Getting top 10 outliers from clusters")
#Trim dataframe to outliers as identified by cluster analysis 
outliers = BankLoans.iloc[(clusters == -1).nonzero()]
outliers
```
