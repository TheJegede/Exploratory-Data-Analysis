### Exercise 3 - Correlation

```python
import pandas as pd
```

```python
SalaryData = pd.read_csv("Salary Data - Ex3.csv")
```

```python
SalaryData
```

#### Covariance Calculations

```python
cov_salary_education = SalaryData['salary'].cov(SalaryData['education'])
cov_salary_prestige = SalaryData['salary'].cov(SalaryData['prestige'])
cov_education_prestige = SalaryData['education'].cov(SalaryData['prestige'])
```

#### Reporting the Results

```python
print(f"Covariance between salary and education: {cov_salary_education}")
print(f"Covariance between salary and prestige: {cov_salary_prestige}")
print(f"Covariance between education and prestige: {cov_education_prestige}")
```

#### Calculating the Pearsonâ€™s correlation coefficients and the p-values

```python
from scipy.stats import pearsonr
```

```python
# Pearson's correlation for salary and education
corr_salary_education, pval_salary_education = pearsonr(SalaryData['salary'], SalaryData['education'])

# Pearson's correlation for salary and prestige
corr_salary_prestige, pval_salary_prestige = pearsonr(SalaryData['salary'], SalaryData['prestige'])

# Pearson's correlation for education and prestige
corr_education_prestige, pval_education_prestige = pearsonr(SalaryData['education'], SalaryData['prestige'])
```

#### Reporting the Results

```python
print(f"Pearson's correlation between salary and education: {corr_salary_education}, p-value: {pval_salary_education}")
print(f"Pearson's correlation between salary and prestige: {corr_salary_prestige}, p-value: {pval_salary_prestige}")
print(f"Pearson's correlation between education and prestige: {corr_education_prestige}, p-value: {pval_education_prestige}")
```
