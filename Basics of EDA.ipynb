### Selecting Rows

#### Rows with salary above 127 or equal

```python
SalaryAbove = Lab1.loc[Lab1["salary"] >= 127, "salary"].tolist()
```

```python
SalaryAbove
```

#### Rows with salaries above the average salary

```python
AvgSalary = sum(Lab1["salary"])/len(Lab1["salary"])
```

```python
AvgSalary
```

```python
SalaryAboveAvg = Lab1.loc[Lab1["salary"] >= AvgSalary, "salary"].tolist()
```
