### Calculating the dissimilarity score between two pairs of data objects 

```python
GameData = pd.read_csv("games_Data.csv")
```

```python
GameData
```

```python
# I will create a function that calculates the dissimilarity score
def calculate_dissimilarity(game1, game2, attributes):
    p = len(attributes)
    m = sum(game1[attr] == game2[attr] for attr in attributes)
    d = (p - m) / p
    return d
```

```python
# This are the rows i will need for comparison
witcher_3 = GameData[GameData["Name"] == "The Witcher 3: Wild Hunt"].iloc[0]
mortal_kombat_11 = GameData[GameData["Name"] == "Mortal Kombat 11"].iloc[0]
super_mario_bros = GameData[GameData["Name"] == "Super Mario Bros."].iloc[0]
super_sonic = GameData[GameData["Name"] == "Super Sonic"].iloc[0]
```

```python
attributes = ["Platform", "Genre", "Publisher"]
```

```python
# Now i will calculate the dissimilarity scores
d_witcher_vs_mortal = calculate_dissimilarity(witcher_3, mortal_kombat_11, attributes)
d_mario_vs_sonic = calculate_dissimilarity(super_mario_bros, super_sonic, attributes)
```

```python
# Printing the results
print(f"Dissimilarity (Witcher 3 vs. Mortal Kombat 11): {d_witcher_vs_mortal}")
print(f"Dissimilarity (Super Mario Bros. vs. Super Sonic): {d_mario_vs_sonic}")
```
